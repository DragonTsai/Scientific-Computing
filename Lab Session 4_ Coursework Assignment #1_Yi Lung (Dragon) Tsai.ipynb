### Exercise 18
Write a Python function that implements the quadratic formula for finding the roots of a quadratic function, in such a way that it does not suffer from at least one of the two potential sources of catastrophic cancellation.

Hint: Consult David Goldberg's paper for this, which is mentioned in the slides and available in the Reading Material section on KEATS. Study in particular the section on catastrophic cancellation.

> *Ideas: Quadratic equation is $ax^2 + bx + c$*
>+ *Determinant $D = b^2 - 4ac$*
>+ *$D > 0$ means two real, distinct roots.* 
>
> *$x_1 = \frac{-b + \sqrt{b^2 -4ac}}{2a}$ & $x_2 = \frac{-b - \sqrt{b^2 -4ac}}{2a}$*
>+ *$D = 0$ means two real, identical roots.*
>
> *$x = \frac{-b}{2a}$*
>+ *$D < 0$ means no real roots.*

> *Catastrophic cancellation occurs after adding two nearly equal numbers of opposite sign. Therefore, if $D > 0$, $x_1 = \frac{-b + \sqrt{b^2 -4ac}}{2a}$ x $\frac{-b - \sqrt{b^2 -4ac}}{-b - \sqrt{b^2 -4ac}} = \frac{2c}{-b - \sqrt{b^2 -4ac}}$, and then we can use $\frac{2c}{-b - \sqrt{b^2 -4ac}}$ to calculate the root without causing catastrophic cancellation.*

```python
import math 

def solve_quad(a,b,c):
    # Calculate the discriminant
    D = b*b - 4*a*c
    # Check whether there are real roots or not
    if D < 0:
        return "no real roots"
    elif D == 0:
        x_1 = (-b)/(2*a)
        x_2 = x_1
        return "two real, identical roots :",x_1, x_2 
    else:
        if b > 0:
            x_1 = (2*c)/(-b-math.sqrt(D)) # Use the formula provided above  
            x_2 = (-b - math.sqrt(D))/2*a # There is no catastrophic cancellation in x_2
            return "two real, distinct roots :",x_1, x_2
        
print(solve_quad(1,1e10,4)) # Test the answer 
    
```
